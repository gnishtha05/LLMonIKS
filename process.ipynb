```python
from tqdm.auto import tqdm
import re
```

```python
DEV_NUM_RE = re.compile(r"[\u0966-\u096F]+")  # ०-९

DEV_TO_AR = str.maketrans("०१२३४५६७८९", "0123456789")

def has_numbering(line: str):
    """
    Returns:
      dev_nums: list of Devanagari digit strings (e.g., ['२', '१', '१२'])
      int_nums: list of ints in the same order (e.g., [2, 1, 12])
    """
    dev_nums = DEV_NUM_RE.findall(line)
    int_nums = [int(s.translate(DEV_TO_AR)) for s in dev_nums]
    return dev_nums, int_nums
```

```python
fname = 'gita.txt'

with open(fname, "r", encoding="utf-8") as f:
    for i, line in tqdm(enumerate(f)):
        hindi_num, int_num = has_numbering(line)
        if len(hindi_num) > 1:
            print(f"  Dev: {hindi_num}, Int: {int_num}, index: {i}")
            print(f"  Line: {line}")
```
